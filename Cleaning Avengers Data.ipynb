# Exploring The Data

```python
import pandas as pd

avengers = pd.read_csv("avengers.csv", encoding='latin-1')
avengers.head(5)
```

# Plotting The Data to Visualize It

```python
import matplotlib.pyplot as plt
%matplotlib inline

avengers['Year'].hist()
```

Avengers were not introduced until 1960s. So data that is included before 1960 should be removed

# Filtering Out Bad Data

```python
true_avengers = pd.DataFrame()
true_avengers = avengers[avengers["Year"] > 1959]
true_avengers
```

# Consolidating Deaths

```python
def clean_deaths(row):
    num_deaths = 0
    cols = ["Death1", "Death2", "Death3", "Death4", "Death5"]
    
    for c in cols:
        death = row[c]
        if pd.isnull(death) or death == 'NO':
            continue
        elif death == 'YES':
            num_deaths += 1
    return num_deaths

true_avengers["Deaths"] = true_avengers.apply(clean_deaths, axis=1)
true_avengers
```

# Verifying Years Since Joining

```python
joined_accuracy_count  = int()
correct_joined_years = true_avengers[true_avengers['Years since joining'] == (2015 - true_avengers['Year'])]
print(correct_joined_years.head())

joined_accuracy_count = len(correct_joined_years)
joined_accuracy_count
```
